#### Alumni Count by Education Degree

# Description:

The code generates a bar chart using Matplotlib to visualize the distribution of alumni data based on their education degrees. The DataFrame 'df' is presumed to contain the alumni data.

1. It sets up a figure with dimensions 10x8 inches.

2. It retrieves the value counts of the 'Education Degree' column from the DataFrame and selects the top 15 values. It calculates the total count and percentage of each education degree.

3. Subsequently, it plots a bar chart representing the count of alumni for each education degree. Above each bar, it displays the count and percentage in the format: "<count>(<percentage>%)", providing additional insight into the distribution.

4. The x-axis is labeled as 'Education Degree', the y-axis as 'Count', and the chart is titled 'Alumni Count and Percentage by Education Degree'.

5. Furthermore, a description is added at the top right corner of the chart to provide context for the visualization.

6. Finally, the chart is displayed with tight layout to ensure all labels fit within the figure.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

# Get the value counts
value_counts = df['Education Degree'].value_counts().head(15)

# Calculate the total count
total_count = value_counts.sum()

# Calculate percentages
percentages = (value_counts / total_count) * 100

# Plot the bar chart
ax = value_counts.plot(kind='bar')

# Add count and percentage labels above each bar in vertical alignment
for i, (count, percentage) in enumerate(zip(value_counts, percentages)):
    plt.text(i, count + 1, f"{count}({percentage:.2f}%)", ha='center', va='bottom')
    # plt.text(i, count - 3, f"({percentage:.2f}%)", ha='center', va='top', rotation=90)

# Set labels and title
plt.xlabel('Education Degree')
plt.ylabel('Count')
plt.title('Alumni Count and Percentage by Education Degree')

# Add description at the top right corner
description = "This chart displays the count and percentage of alumni by education Degree."
plt.text(0.85, 0.95, description, transform=ax.transAxes, fontsize=10, ha='right', va='top')

# Show the chart
plt.tight_layout()  # Ensure the labels fit within the figure
plt.show()
```

## **1. Clustering**

This code performs clustering on the alumni dataset using K-Means, where it selects relevant features such as 'Education Major,' 'Location City,' and 'Company Industry Name.' It then applies K-Means clustering with three clusters and calculates the Silhouette Score to evaluate the clustering quality. The 'Cluster' column in the dataset contains cluster labels assigned to each alumni based on the selected features.


```python
from google.colab import drive

drive.mount('/content/drive')


import pandas as pd

df = pd.read_csv('/content/drive/My Drive/alumni_dataset_1187.csv')
```

This code snippet performs clustering analysis on an alumni dataset using K-Means algorithm. Here's a breakdown of the steps:

1. **Loading the Dataset**: The code assumes that the alumni dataset is already loaded into a pandas DataFrame named `df`.

2. **Feature Selection**: Relevant features for clustering are selected, including 'Education Major', 'Location City', and 'Company Industry Name'.

3. **Preprocessing**:
   - Numerical and categorical features are identified.
   - For numerical features, standard scaling is applied using `StandardScaler`.
   - For categorical features, one-hot encoding is applied using `OneHotEncoder`.

4. **Column Transformation**: A `ColumnTransformer` is created to apply the preprocessing steps to numerical and categorical features separately.

5. **Pipeline Creation**: A pipeline is created to combine the preprocessing steps with the K-Means clustering algorithm. The pipeline consists of:
   - Preprocessor: Applies scaling and one-hot encoding to the features.
   - Clusterer: Uses K-Means algorithm for clustering. The number of clusters is set to 3, but it can be adjusted as needed.

6. **Fitting the Pipeline**: The pipeline is fitted to the selected features.

7. **Adding Cluster Labels**: Cluster labels obtained from the K-Means algorithm are added to the original dataset as a new column named 'Cluster'.

8. **Silhouette Score Calculation**: The silhouette score is calculated based on the numerical features used for clustering. It evaluates the quality of the clusters obtained. Higher silhouette scores indicate better-defined clusters.

9. **Output**: The silhouette score is printed to the console, providing an indication of how well the data points are clustered. Additionally, the 'Cluster' column in the dataset now contains cluster labels assigned to each alumni based on the selected features.



The silhouette score is a measure used to evaluate the quality of clusters created by a clustering algorithm. It quantifies how well-separated the clusters are, with values ranging from -1 to 1.

- A silhouette score close to +1 indicates that the data point is well-clustered and lies far away from neighboring clusters.
- A score of 0 indicates that the data point is on or very close to the decision boundary between two neighboring clusters.
- A score close to -1 indicates that the data point may have been assigned to the wrong cluster.

In general, higher silhouette scores indicate better cluster separation. However, it's essential to interpret silhouette scores in the context of your specific dataset and problem.


```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import silhouette_score

# Load your alumni dataset
# data = pd.read_csv(df)

# Select the relevant features for clustering (e.g., 'Education Major', 'Location City', 'Company Industry Name')
selected_features_for_clustering = df[['Education Major', 'Location City', 'Company Industry Name']]

# Preprocessing for numerical and categorical features
numerical_features = selected_features_for_clustering.select_dtypes(include=['int64', 'float64']).columns
categorical_features = selected_features_for_clustering.select_dtypes(include=['object']).columns

# Create a preprocessing pipeline for numerical and categorical features
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Combine preprocessing with clustering using K-Means
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('clusterer', KMeans(n_clusters=3, random_state=42))])  # You can specify the number of clusters

# Fit the pipeline to your data
pipeline.fit(selected_features_for_clustering)

# Add cluster labels to the original dataset
df['Cluster'] = pipeline.named_steps['clusterer'].labels_

# Calculate Silhouette Score based on the numerical features used for clustering
silhouette_avg = silhouette_score(pipeline.transform(selected_features_for_clustering), df['Cluster'])

print(f'Silhouette Score: {silhouette_avg}')

# Now, the 'Cluster' column in your dataset contains the cluster labels assigned to each alumni based on mixed features.
```

### **Analyze employment titles using NLP techniques to identify common keywords, skills, or qualifications associated with specific job roles.**

This Python code preprocesses employment titles from a dataset by tokenizing, lowercasing, and removing stopwords. It then performs topic modeling using Latent Dirichlet Allocation (LDA) and visualizes the distribution of topics for a specific job role, allowing you to analyze the most relevant topics associated with that role.

Please replace the "job_role" variable with the specific job title you want to analyze to see its topic distribution.

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
import nltk
nltk.download('punkt')

# # Load your dataset
# data = pd.read_csv(url)

# Preprocess the employment titles
stop_words = set(stopwords.words('english'))  # NLTK stopwords

# Tokenize and preprocess the titles
def preprocess_title(title):
    tokens = nltk.word_tokenize(title.lower())  # Tokenization and lowercase
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]  # Remove non-alphabetic and stopwords
    return " ".join(tokens)

df['Processed Title'] = df['Employment Title'].apply(preprocess_title)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=1000, max_df=0.85)
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Processed Title'])

# Topic Modeling using LDA
num_topics = 5  # You can adjust this number
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(tfidf_matrix)

# Get top keywords for each topic
def get_top_keywords(model, feature_names, n_words=10):
    topic_keywords = []
    for topic_idx, topic in enumerate(model.components_):
        top_keywords_idx = topic.argsort()[:-n_words - 1:-1]
        top_keywords = [feature_names[i] for i in top_keywords_idx]
        topic_keywords.append(top_keywords)
    return topic_keywords

feature_names = tfidf_vectorizer.get_feature_names_out()
topic_keywords = get_top_keywords(lda, feature_names)

# Visualize top keywords for each topic
for i, keywords in enumerate(topic_keywords):
    print(f"Keyword Topic {i + 1}: {', '.join(keywords)}")

# Plot the distribution of topics
topic_distribution = lda.transform(tfidf_matrix)
topic_distribution_df = pd.DataFrame(topic_distribution, columns=[f'Topic {i}' for i in range(1, num_topics + 1)])
df = pd.concat([df, topic_distribution_df], axis=1)




# --------------------------------Replace with the job role you want to analyze-----------------------------------------
job_role = "Program Manager"



# Plot the distribution of topics for a specific job role
job_data = df[df['Employment Title'] == job_role]
topic_weights = job_data.iloc[:, -num_topics:]  # Extract topic weights for the job role
topic_weights = topic_weights.mean()  # Calculate mean weights across job instances
plt.bar(range(1, num_topics + 1), topic_weights)
plt.xlabel('Keyword Topic')
plt.ylabel('Topic Weight')
plt.title(f'Topic Distribution for {job_role}')
plt.show()
```
